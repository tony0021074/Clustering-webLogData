```python
import pandas as pd
```

## Read raw data and convert it to list

```python
lists = []

with open("../input/anonymous-msweb.data", "r") as f:
    for line in f:
        if line[0] == "C":
            key = int(line[-6:-1])
            lists.append([key,[]]) #[[10001]]
        if line[0] == "V" and key > 0:
            read_value = int(line[2:6])
            lists[-1][1].append(read_value)
lists
```

## Study data

```python
print('Number of visitor: ' + str(len(lists)))
```

```python
def find_duplicate_visitor(lists):
    isDuplicate = False
    for i, visitor1 in enumerate(lists):
        for visitor2 in lists[i+1:]:
            if visitor1[0]==visitor2[0]:
                isDuplicate = True
                break
        if isDuplicate == True:
            break
    if isDuplicate == True:
        print("Duplicate Visitor Found")
    else:
        print("Duplicate Visitor Not Found")
        
            
find_duplicate_visitor(lists)
```

```python
def find_duplicate_area_in_visitor(lists):
    isDuplicate = False
    for user in lists:
        visits = user[1]
        for i, visit1 in enumerate(visits):
            for visit2 in visits[i+1:]:
                if visit1==visit2:
                    isDuplicate = True
                    break
            if isDuplicate == True:
                break
    if isDuplicate == True:
        print("Duplicate Area Found")
    else:
        print("Duplicate Area Not Found")
        
            
find_duplicate_area_in_visitor(lists)
```
